```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, median_absolute_error, r2_score, mean_absolute_percentage_error
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
output = pd.read_csv('output.csv')
output
```

```python
print("seldon MAE", mean_absolute_error(output['ata'], output['seldon_eta']))
print("snapp MAE", mean_absolute_error(output['ata'], output['snapp_eta']))

print("seldon MSE", mean_squared_error(output['ata'], output['seldon_eta']))
print("snapp MSE", mean_squared_error(output['ata'], output['snapp_eta']))

print("seldon explained_variance_score", explained_variance_score(output['ata'], output['seldon_eta']))
print("snapp explained_variance_score", explained_variance_score(output['ata'], output['snapp_eta']))

print("seldon median_absolute_error", median_absolute_error(output['ata'], output['seldon_eta']))
print("snapp median_absolute_error", median_absolute_error(output['ata'], output['snapp_eta']))

print("seldon r2_score", r2_score(output['ata'], output['seldon_eta']))
print("snapp r2_score", r2_score(output['ata'], output['snapp_eta']))

print("seldon MAPE", mean_absolute_percentage_error(output['ata'], output['seldon_eta']))
print("snapp MAPE", mean_absolute_percentage_error(output['ata'], output['snapp_eta']))
```
